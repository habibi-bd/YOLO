.yaml file does NOT contain the actual images, videos, or other media files themselves.
Instead, it contains metadata and configuration information about where those files are
located and how to interpret them.

# train: ../coco8/images/train2017   # path to training images folder
# val: ../coco8/images/val2017       # path to validation images folder

# nc: 8                             # number of classes
# names:                           # class names (labels)
#   0: person
#   1: bicycle
#   2: car
#   3: motorcycle
#   4: airplane
#   5: bus
#   6: train
#   7: truck
